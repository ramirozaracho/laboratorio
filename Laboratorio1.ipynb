# Primer laboratorio de Analisis de Datos

## ETL (Extraccion, Transformacion y Carga)

### 1. Extracción

```python
from google.colab import drive
import pandas as pd
import chardet

# Montamos google drive para acceder al archivo
drive.mount('/content/drive', force_remount=True)

ruta_archivo = '/content/drive/MyDrive/Visas_h-1b.csv'
with open(ruta_archivo, 'rb') as f:
    result = chardet.detect(f.read())


try:
  df = pd.read_csv(ruta_archivo, encoding=result['encoding'])
  print('extraccion de datos exitosa')
except FileNotFoundError:
  print(f'El archivo no se encuentra en la ruta: {ruta_archivo}')
```

### 2. Transformación

```python
# Muestra las primeras 5 filas del DataFrame
df.head()
```

```python
# Renombrar columnas para traducirlas al español y entender mejor
df.rename(columns ={
    'CASE_NUMBER' : 'Numero_de_caso',
    'CASE_STATUS' : 'Estado_del_caso',
    'EMPLOYER_NAME' : 'Nombre_de_empresa',
    'SOC_NAME' : 'Titulo',
    'SOC_CODE' : 'Codigo_de_titulo',
    'JOB_TITLE' : 'Ocupacion',
    'FULL_TIME_POSITION' : 'Tiempo_completo',
    'PREVAILING_WAGE' : 'Salario_previsto',
    'WORKSITE_CITY' : 'Ciudad',
    'WORKSITE_STATE_ABB' : 'Estado_abreviado',
    'YEAR' : 'Año',
    'WORKSITE_STATE_FULL' : 'Estado',
    'WORKSITE' : 'Lugar de trabajo',
}, inplace=True)
```

```python
# Resumen de la informacion del DataFrame
df.info()
```

```python
# Se renombran las filas de las columnas para traducirlas al español
df['Estado_del_caso'] = df['Estado_del_caso'].str.replace('CERTIFIED', 'CERTIFICADA')
df['Estado_del_caso'] = df['Estado_del_caso'].str.replace('WITHDRAWN', 'RETIRADA')
df['Estado_del_caso'] = df['Estado_del_caso'].str.replace('CERTIFIED-WITHDRAWN', 'CERTIFICADO-DENEGADO')
df['Estado_del_caso'] = df['Estado_del_caso'].str.replace('DENIED', 'DENEGADA')

df['Titulo'] = df['Titulo'].str.replace('COMPUTER SYSTEMS ANALYSTS', 'ANALISTA DE SISTEMAS')
df['Titulo'] = df['Titulo'].str.replace('COMPUTER OCCUPATIONS, ALL OTHER*', 'OCUPACIONES DE ORDENADOR, ENTRE OTROS')
df['Titulo'] = df['Titulo'].str.replace('SOFTWARE DEVELOPERS, APPLICATIONS', 'DESAROLLADOR DE SOFTWARE')
df['Titulo'] = df['Titulo'].str.replace('FILM AND VIDEO EDITORS', 'EDITOR DE CINE Y VIDEO')
df['Titulo'] = df['Titulo'].str.replace('NETWORK AND COMPUTER SYSTEMS ADMINISTRATORS', 'ADMINISTRADOR DE REDES Y SISTEMAS')
df['Titulo'] = df['Titulo'].str.replace('COMPUTER PROGRAMMERS', 'PROGRAMADOR DE COMPUTADORA')
df['Titulo'] = df['Titulo'].str.replace('FINANCIAL ANALYSTS', 'ANALISTA FINANCIERO')
df['Titulo'] = df['Titulo'].str.replace('OPERATIONS RESEARCH ANALYSTS', 'ANALISTA DE INVESTIGACION')

df['Ocupacion'] = df['Ocupacion'].str.replace('SENIOR BUSINESS ANALYST', 'ANALISTA DE NEGOCIOS SENIOR')
df['Ocupacion'] = df['Ocupacion'].str.replace('BQUALITY ASSURANCE ANALYST', 'ANALISTA DE CONTROL DE CALIDAD')
df['Ocupacion'] = df['Ocupacion'].str.replace('SR. SOFTWARE DEVELOPER', 'DESAROLLADOR DE SOFTWARE SENIOR')
df['Ocupacion'] = df['Ocupacion'].str.replace('VIDEO EDITOR', 'EDITOR DE VIDEO')
df['Ocupacion'] = df['Ocupacion'].str.replace('SYSTEMS ADMINISTRATOR', 'ADMINISTRADOR DE SISTEMAS')
```

### 3. EDA (Analisis Exploratorio de Datos)

```python
# ¿Que estados, ciudades o empresas ofrecen la mayor cantidad de visas H1-B?
df_aprobadas = df.groupby('Estado')['Nombre_de_empresa'].value_counts(ascending=False)

print('La empresa que mayor cantidad de visas ofrece es Infosys Limited')

print(df_aprobadas)
```

```python
# ¿En que año y de que ocupacion se rechazaron mas visas?

# Filtra solo por los casos denegados
df_rechazadas = df[df['Estado_del_caso'] == 'DENEGADA']

# Cuenta cuantos denegaciones hubo por año y ocupacion
rechazos_por_año_ocupacion = (df_rechazadas.groupby(['Año', 'Ocupacion']).size().reset_index(name='Cantidad_rechazos'))

# Busca la fila con el mayor numero de rechazos
mayor_rechazo = rechazos_por_año_ocupacion.loc[rechazos_por_año_ocupacion['Cantidad_rechazos'].idxmax()]

print(f"Año con más rechazos:", mayor_rechazo['Año'])
print(f"Ocupación con más rechazos:", mayor_rechazo['Ocupacion'])
print(f"Cantidad de rechazos:", mayor_rechazo['Cantidad_rechazos'])
```

```python
# Según tu perfil laboral, ¿en qué ciudad deberías estar para tener más oportunidades?

# Agrupa por ciudad la cantidad de oportunidades
oportunidades_por_ciudad = (df.groupby('Ciudad').size().reset_index(name='Cantidad_oportunidades'))

# Ordena de mayor a menor cantidad
mejores_diez = oportunidades_por_ciudad.sort_values('Cantidad_oportunidades',ascending=False)

# Selecciona la primer ciudad de la lista
mejor_ciudad = mejores_diez.iloc[0]['Ciudad']

print(mejores_diez.head(10))
print(f'La ciudad con más oportunidades laborales es: {mejor_ciudad}')
```
